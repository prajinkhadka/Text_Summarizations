```python
import bs4 as bs
```

```python
import urllib.request
```

```python
#getting data
source  = urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming')
```

```python
source
```

```python
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
source
```

```python
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
    
```

```python
text
```

```python
#pre processing the text

import re
```

```python
text = re.sub(r'\[[0-9]*\]', ' ', text) #remoinbg [ brackets]
```

```python
text = re.sub(r'\s+', ' ', text) #removing the extra spaces
```

```python
clean_text = text.lower()
```

```python
clean_text = re.sub(r'\W'," ", clean_text)
clean_text = re.sub(r'\d', " ", clean_text)
clean_text = re.sub(r'\s+', " ",clean_text)
```

```python
clean_text
```

```python
#tokenizing into sentences
import nltk
nltk.download('stopwords')
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
len(sentences)
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
stop_words
```

```python
#creating histogram
word2count ={}
for word in nltk.word_tokenize(clean_text):
    if word not in stop_words:
        if word not in word2count.keys():
            word2count[word] =1
        else:
            word2count[word] += 1
```

```python
word2count
```

```python
#creating weighted histogram

for key in word2count.keys():
    word2count[key] = word2count[key] /max(word2count.values())
```

```python
word2count
```

```python
#creating sentences scores

sent2score = {}
for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word2count.keys():
            if len(sentence.split(' ')) <30:
                if sentence not in sent2score.keys():
                    sent2score[sentence] = word2count[word]
                else:
                    sent2score[sentence] += word2count[word]
```

```python
(sent2score)
```

```python
import heapq
```

```python
best_sentences = heapq.nlargest(25,sent2score, key = sent2score.get)
```

```python
best_sentences
```

```python
print('--------------------------------------------------------------')
for sentence in best_sentences:
    print(sentence)
```
